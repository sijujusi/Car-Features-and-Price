# Introduction

Cars are popular mode of transport. There are many companies who manufacture them and there are several features which can influence their price. The goal of this project is to find features which can potentially impact the price of a car. 

The data come from the Kaggle dataset Car Features and MSRP of Kaggle. It describes almost 12000 car models sold in the USA between 1990 and 2018 with the market price (new or used) and some features. This project is performed to carry out some data manipulations and carry out exploratory data analysis .

### Description of Data
Dataset from Kaggle : https://www.kaggle.com/CooperUnion/cardataset

This dataset contains more of 15 features of the car such as Make, Model, Engine Fuel Type, Engine HP, Transmission Type, highway MPG, city MPG, vehicle size,  and many more.

The values selected were based upon the boxplots. Other method is to use the interquartile range. 

## Example of filtering

```python
df_before_2000_filtered = df.loc[((df['Year']<=2000) & (df['MSRP']< 10000))]
df_after_2000_filtered = df.loc[((df['Year']>2000) & (df['MSRP']< 500000))]
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x="Year", y ='MSRP', data=df_after_2000_filtered)
plt.title("Box plot for filtered data after year 2000",fontsize=25)
plt.show()
```

We can see we have eliminated unrealistic outliers from the datasets. However still there are lot of outliers which might be becuase of differenes in prices of the car due to different in their features such as model, make, size, horse power, etc.

# Filtering and Cleaning Data

```python
# Dropping irrelevant columns
df1 = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
```

```python
df1 = df1.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df1.head(5)
```

```python
df1.shape
```

# Deleting duplicate data


```python
duplicate_rows_df = df1[df1.duplicated()]
print("number of dupicate rows:", duplicate_rows_df.shape)
```

```python
df1.count()
```

```python
df1 = df1.drop_duplicates()
df1.head(5)
```

```python
df1.describe()
```

```python
fig, ax =plt.subplots(2,2, figsize=(15,10))
sns.boxplot(x='Price', data=df1,ax=ax[0,0])
sns.boxplot(x='HP', data=df1,ax=ax[0,1])
sns.boxplot(x='MPG-H', data=df1,ax=ax[1,0])
sns.boxplot(x='MPG-C', data=df1,ax=ax[1,1])
```

```python
sns.boxplot(x=df1['Cylinders'])
```

   ## Minimization of outliers by using Interquartile range 

```python
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1  #interquantile range 
print(IQR)
```

```python
df1 = df1[~((df1 < (Q1 - 1.5 * IQR)) |(df1 > (Q3 + 1.5 * IQR))).any(axis=1)]
df1.shape
```

```python
plt.figure(figsize=(10,5))
c = df1.corr()
sns.heatmap(c,cmap='BrBG', annot= True)
plt.title('Heatmap to show correlation between different features',fontsize=20 )
plt.show()
c
```

```python
sns.scatterplot(x='HP', y = 'Price', data = df1)
plt.title('Scatter plot Engine HP vs MSRP',fontsize=20 )
plt.show()

```

Clearly, the MSRP (Manufactur Suggested Retail Price) increases with the increase in the Engine HP.
